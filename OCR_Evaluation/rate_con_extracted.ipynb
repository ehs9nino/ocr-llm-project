```python
import os
from paddleocr import PaddleOCR
from PIL import Image
import pandas as pd
```

```python
# Define input/output paths
image_dir = '/root/pytesseract/testData'
output_dir = '/root/pytesseract/ocr_outputs'
os.makedirs(output_dir, exist_ok=True)
```

```python
# Init
ocr = PaddleOCR(use_angle_cls=True, lang='en')
image_dir = '/root/pytesseract/testData'
results = []

# Extract OCR and store
for img_name in sorted(os.listdir(image_dir)):
    if img_name.lower().endswith('.jpg'):
        img_path = os.path.join(image_dir, img_name)
        result = ocr.ocr(img_path, cls=True)
        text = "\n".join([line[1][0] for line in result[0]])
        results.append({'Image': img_name, 'Text': text})

# Save to CSV
df = pd.DataFrame(results)
csv_path = '/root/pytesseract/ocr_outputs/ocr_combined.csv'
df.to_csv(csv_path, index=False, encoding='utf-8')

print("✅ OCR results saved to:", csv_path)
```
